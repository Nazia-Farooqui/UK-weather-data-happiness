```python
import pandas as pd
import metdata
from matplotlib import pyplot as plt
```

```python
metdata.load_all()
```

```python
def find_first_year(stations_list, all_stations_data):
    years_list = []
    if len(all_stations_data) > 0:
        for station in stations_list:
            if station in all_stations_data:
                station_data = all_stations_data[station]
                for record in station_data.itertuples(index = False):
                    if hasattr(record, 'yyyy'):
                        year = record.yyyy
                        month = record.mm
                        if int(month) == 1:
                            years_list.append(int(year))
                            break  #stop iterating for current station once a January is matched
        return years_list
    else:
        return None
```

```python
stations_list = metdata.all_names()
```

```python
all_stations_data = metdata.dfs
```

```python
years_list = find_first_year(stations_list, all_stations_data)
print("years_list =", years_list)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import silhouette_score
from sklearn.cluster import DBSCAN
```

```python
### *** Scatterplots above suggests that TMIN & AF have almost linear relationship, so default options are TMAX/RAIN/SUN ***
```

```python
# Select the features you want to use for clustering
features = ['TMAX', 'RAIN', 'SUN']

# Create the array that will contain the data - include only the selected features
data = df[features].values

# Perform K-means clustering
num_clusters = 3
n_init_value = 10  # Set the value of n_init explicitly

kmeans = KMeans(n_clusters=num_clusters, n_init=n_init_value)
kmeans.fit(data)

#get cluster labels and cluster centres
labels = kmeans.labels_
cluster_centres = kmeans.cluster_centers_

# add cluster labels to the DataFrame
df_cluster = df.copy()
df_cluster['Cluster'] = labels

df_cluster
```

```python
# Create a 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Define colors for each cluster
colors = ['red', 'green', 'blue']

# Plot data points with different colors for each cluster
for cluster_id in range(num_clusters):
    cluster_data = df_cluster[df_cluster['Cluster'] == cluster_id]
    ax.scatter(cluster_data['TMAX'], cluster_data['RAIN'], cluster_data['SUN'], c=colors[cluster_id], label=f'Cluster {cluster_id}')

# Set labels for the axes
ax.set_xlabel('TMAX')
ax.set_ylabel('RAIN')
ax.set_zlabel('SUN')

# Set a title and legend
plt.title('Weather Data Clustering')
ax.legend()

# Show the plot
plt.show()
```

```python
# Select the features for clustering
features = ['TMAX', 'RAIN', 'SUN']
data = df[features].values

# Perform DBSCAN clustering
db = DBSCAN(eps=0.5, min_samples=5)
labels = db.fit_predict(data)

# Plot the clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.xlabel('TMAX')
plt.ylabel('RAIN')
plt.title('DBSCAN Clustering')
plt.colorbar()
plt.show()
```

### ***Each station is assigned with the Cluster data point. The cluster labels will be integers from 0 to K-1, where K is the number of clusters you specified. Each integer represents a specific cluster. However, each station could be assigned one or more cluster data point***

### ***The silhouette score is a measure of how well-defined the clusters are in a clustering algorithm. It quantifies the separation and compactness of the clusters. The silhouette score ranges from -1 to 1, where:

### A score close to 1 indicates that the samples in one cluster are well separated from the samples in other clusters. A score close to 0 indicates that the samples are close to the decision boundary between two neighboring clusters. A score close to -1 indicates that the samples are assigned to the wrong clusters.***

```python
labels = df_cluster['Cluster']
```

```python
# Calculate the silhouette score
silhouette_avg = silhouette_score(data, labels)

# Print the silhouette score
print("Silhouette Score:", silhouette_avg)
```
